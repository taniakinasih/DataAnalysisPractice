## Figure 1

```python
import numpy as np
from matplotlib import pyplot as plt

a   = np.load( 'data-fig1.npy' )

plt.figure()
plt.plot (a, c="black", marker='.', mew=4)
plt.xlabel ("Time")
plt.ylabel ("Value")
plt.show()
```

## Figure 2

```python
import numpy as np
from matplotlib import pyplot as plt

b   = np.load( 'data-fig2.npy' )

plt.figure()

plt.plot(b[:, 0], color='r', label="Red")
plt.plot(b[:, 1], color='g', label="Green")
plt.plot(b[:, 2], color='b', label="Blue")
plt.xlabel ("X Value")
plt.ylabel ("Y Value")
plt.legend()

plt.show()
```

## Figure 4

```python
import numpy as np
from matplotlib import pyplot as plt

d   = np.load( 'data-fig4.npy' )
categories = ['A', 'B', 'C', 'D', 'E']

plt.figure()
plt.bar(categories, d, color="silver")
plt.xlabel ("Category")
plt.ylabel ("Value")
plt.show()
```
